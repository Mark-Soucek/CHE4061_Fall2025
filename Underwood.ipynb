```python
alphas = np.array([3., 2., 1., 0.5])
zs = np.full(4, 0.25)
q=0.5
```

```python
def uw(phi):
  return (1-q)-np.sum(alphas*zs/(alphas-phi))
```

```python
phis = np.linspace(0.4, 3.1, 500)
uws =  [uw(phi) for phi in phis]
```

```python
fig = make_subplots()
fig.add_scatter(x = phis, y = uws, mode = 'markers')
```

```python
#1.38477 = phi between active phis
xd_mol = np.array([0.25, 0.2, 0.05, 0])
xd = xd_mol/np.sum(xd_mol)
V = np.sum(alphas*xd/(alphas - 1.38477))
```
