```python
#Smith and Corripio FOPTD
Gp = 1.5/(2.2*s+1)/(1.7*s+1)/(3.5*s+1)/(4*s+1)
tplot, hplot = sim(Gp, lambda t: 1, N = 30, dt = 1.5)
y = np.random.normal(loc = hplot, scale = 0.01)
fig = make_subplots()
fig.add_scatter(x = tplot, y = y, mode= 'markers')
```

```python
K = 1.5
tau = 8
theta = 4
```

```python
def sumsqerr(v):
  K, tau, theta = v
  y_model = np.where(tplot < theta, 0, K*(1-np.exp(-(tplot-theta)/tau)))
  return np.sum((y-y_model)**2)
```

```python
K_fit, tau_fit, theta_fit = sp.optimize.minimize(sumsqerr, [K, tau, theta]).x
K_fit
```

```python
fig = make_subplots()
fig.add_scatter(x = tplot, y = y, mode= 'markers')
fig.add_scatter(x = tplot, y = np.where(tplot < theta, 0, K_fit*(1-np.exp(-(tplot-theta_fit)/tau_fit))), mode= 'lines')
```

```python
PA, PB = 0.586, -0.916
IA, IB = 1.03, -0.165
Y = PA*(theta_fit/tau_fit)**PB
Kc = Y/K_fit
taui = tau_fit/(IA + IB*theta_fit/tau_fit)
Gc = Kc*(1 + 1/(taui*s))
Gcl = Gc*Gp/(1+Gc*Gp)
```

```python
t, y = sim(Gcl, lambda t: 1, N = 1000, dt = 0.1)
fig = make_subplots()
fig.add_scatter(x = t, y = y, mode= 'lines')
```

```python
fig, mag, angle = bode(Gc*Gp)
fig.show()
```
