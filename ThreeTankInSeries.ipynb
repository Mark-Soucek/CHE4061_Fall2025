```python
import numpy as np
from plotly.subplots import make_subplots
from scipy.integrate import solve_ivp
```

```python
A = 1.2 #m2 area of tank
Aout = 0.2 #m2 area of flow line
g=9.81 #m/s2
qin_ss = 0.5 #m3/s flow rate of inlet at ss
h_ss = ((qin_ss/Aout)**2)/(2*g) #height

h3sp = 1.1*h_ss  #setpoint for liquid level of 3rd tank

Kc=5.  #Controller gain

def rhs(t, v):
    h1, h2, h3 = v
    qin0 = qin_ss * 1.1 + Kc * (h3sp - h3)  #Proportional only controller P-controller
    q1, q2, q3 = Aout*np.sqrt(2*g*h1), Aout*np.sqrt(2*g*h2), Aout*np.sqrt(2*g*h3)
    return np.array([(qin0 - q1)/A, (q1-q2)/A, (q2-q3)/A])
```

```python
tend= 20.
res=solve_ivp(rhs, (0, tend), [h_ss]*3, dense_output=True, method='Radau')
res
```

```python
tplot = np.linspace(0, tend, 100)
h1, h2, h3 =res.sol(tplot)
fig=make_subplots()
fig.add_scatter(x=tplot, y=h1, mode='lines', name='h1')
fig.add_scatter(x=tplot, y=h2, mode='lines', name='h2')
fig.add_scatter(x=tplot, y=h3, mode='lines', name='h3')
fig.update_layout(width=600, height=400, template='plotly_dark')
```
