```python
import numpy as np
import scipy as sp
from plotly.subplots import make_subplots
import plotly.io as pio
pio.templates.default = "plotly_dark"

import jax
import jax.numpy as jnp
jax.config.update("jax_enable_x64", True)
```

#Von Neumann

```python
alpha = 1.  #m2/s
dx = 0.25
dt = 0.01

r = dt*alpha/(dx**2)
N=100
```

```python
r=0.4999999
```

```python
T=np.full(N,20.)
T[0]=50.
T[-1]=100
Tprofiles = [T.copy()]
```

```python
for i in range(20000):
    T[1:-1] = T[1:-1] + r*(T[:-2] -2*T[1:-1] + T[2:] )
    Tprofiles.append(T.copy())
```

```python
fig = make_subplots()
for k in Tprofiles[::200]:
    fig.add_scatter(x=np.arange(N), y= k)
fig.show()
```

```python
N=100
T=np.full(N, 20.)
T[0]=50
T[-1]=80
r=100
ab = np.stack([np.full(N-2,-r), np.full(N-2,1+2*r), np.full(N-2,-r)])
Tprofiles = [T.copy()]
```

```python
for i in range(1000):
  b = r*T[:-2] + (1-2*r)*(T[1:-1])+r*T[2:]
  b[0] += r*T[0]
  b[-1] += r*T[-1]
  T[1:-1]=sp.linalg.solve_banded((1,1), ab, b)
  Tprofiles.append(T.copy())
```

```python
fig = make_subplots()
for k in Tprofiles[::100]:
    fig.add_scatter(x=np.arange(N), y= k)
fig.show()
```
